```python
# ler a carteira e funções do programa
from matplotlib import pyplot as plt

import pandas as pd
import numpy as np
import riskwallet_functions as risk

df = pd.read_excel("/Users/gabrielassuncao/Desktop/alta_vol.xlsx")


def colocar_qtd(qtd, stocklist):
    qtd_stocks = np.zeros((len(stocklist)))
    for i in range(len(qtd_stocks)):
        qtd_stocks[i] = qtd


def calcular_pesos(qte_positions):
    weights = qte_positions / sum(qte_positions)
    return(weights)


def performance_carteira(wallet_returns):
    cota = [1]
    for i in range(len(wallet_returns)):
        cota.append(cota[i] * (wallet_returns[i] + 1))
    return cota


def volatilidade_rollingwindow(wallet_returns, window):
    vol = []
    for i in range(window, len(wallet_returns)):
        vol.append( np.std(wallet_returns[(i-window):i]) * np.sqrt(252) )
    return vol


df.head()
```

```python
# performance last 252du, volatilidade last 252du, var last 252du
window = 252
stocklist = np.array(df['stocks'])

price_data, date_data = risk.get_data("2024-01-22", window, stocklist)
return_data = risk.get_returns(price_data)
qte_data = np.zeros((len(stocklist)))
for i in range(len(qte_data)):
    qte_data[i] = 1000.0
pesos_data = calcular_pesos(qte_data)
wallet_returns = risk.hist_wallet_returns(pesos_data, return_data)

print(stocklist)
print(return_data.shape)
print(wallet_returns.shape)
print(" ")

# Ativos volatilidade últimos 21 dias/ Retorno
mediana = np.median(wallet_returns)
media = np.mean(wallet_returns)
volatilidade = np.std(wallet_returns) * np.sqrt(252)
var = risk.calculate_var(wallet_returns, 0.99)

# ibovespa
stock_ibov, date_ibov = risk.get_data("2024-01-22", window, ['BOVA11'])
stock_return = risk.get_returns(stock_ibov)
stock_return = np.squeeze(stock_return)
correl_ibov = np.corrcoef(wallet_returns, stock_return)[0, 1]

# small11
stock_smal, date_smal = risk.get_data("2024-01-22", window, ['SMAL11'])
smal_return = risk.get_returns(stock_smal)
smal_return = np.squeeze(smal_return)
correl_smal = np.corrcoef(wallet_returns, smal_return)[0, 1]

print(f"Com a janela de {window} dias")
print("Correl com Ibov e Correl com Smal: %.2f %.2f" % (correl_ibov, correl_smal))
print("Mediana, Media, Volatilidade, Var : %.4f%% %.4f%% %.2f%% %.2f%%" % (mediana * 100, media * 100,
                                                                           volatilidade * 100, var * 100))
```

```python
# Ativos volatilidade últimos 21 dias/ Retorno
# performance last 252du, volatilidade last 252du, var last 252du
stocklist = np.array(df['stocks'])

price_data, date_data = risk.get_data("2024-01-22", 252, stocklist)
return_data = risk.get_returns(price_data)
qte_data = np.zeros((len(stocklist)))
for i in range(len(qte_data)):
    qte_data[i] = 1000.0
pesos_data = calcular_pesos(qte_data)
wallet_returns = risk.hist_wallet_returns(pesos_data, return_data)

print(stocklist)
print(return_data.shape)
print(wallet_returns.shape)
print(" ")

mediana = np.median(wallet_returns)
media = np.mean(wallet_returns)
volatilidade = np.std(wallet_returns) * np.sqrt(252)
var = risk.calculate_var(wallet_returns, 0.99)

print("Mediana, Media, Volatilidade, Var : %.4f%% %.4f%% %.2f%% %.2f%%" % (mediana * 100, media * 100,
                                                                           volatilidade * 100, var * 100))
```

```python
# performance ultimos 252 dias
cota_carteira = performance_carteira(wallet_returns)
cota_carteira = np.array(cota_carteira)

print("Retorno wallet: %.2f%% " % ((cota_carteira[-1] / cota_carteira[0] - 1) * 100) )
print("Volatilidade carteira: %.2f%% " % (np.std(wallet_returns) * np.sqrt(252) * 100) )

plt.plot(cota_carteira)
plt.title(f"Performance Carteira vol ultimos {len(wallet_returns)} ")

plt.xlabel("Dias - Histórico")
plt.ylabel("Cota Carteira")
plt.show()
```
